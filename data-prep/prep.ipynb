```python
builder = FeatureBuilder()

builder.add_feature(InningsRuns("innings_runs")) \
       .add_feature(InningsBalls("innings_balls")) \
       .add_feature(InningsTotalRuns("innings_final_runs")) \
       .add_feature(InningsTotalBalls("innings_final_balls")) \
       .add_feature(InningsWickets("innings_wickets")) \
       .add_feature(InningsTotalWickets("innings_final_wickets")) \
       .add_feature(BatsmanScore("batsman_score")) \
       .add_feature(BatsmanBalls("batsman_balls")) \
       .add_feature(BatsmanTotalScore("batsman_final_score")) \
       .add_feature(BatsmanTotalBalls("batsman_total_balls")) \
       .add_feature(BowlerConceded("bowler_conceded")) \
       .add_feature(BowlerBalls("bowler_balls")) \
       .add_feature(BowlerTotalConceded("bowler_final_conceded")) \
       .add_feature(BowlerTotalBalls("bowler_total_balls")) \
       .add_feature(InningsWickets("bowler_wickets")) \
       .add_feature(InningsTotalWickets("bowler_final_wickets")) \
       .add_feature(ChaseTarget("target")) \
       .apply(df)
```
